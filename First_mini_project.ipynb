```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3= col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
Irprice,urprice=remove_outlier(df['Price'])
df ['Price']=np.where(df['Price']<Irprice,Irprice,df['Price'])
df ['Price']=np.where(df['Price']>urprice,urprice,df['Price'])

IrHP,urHP=remove_outlier(df['HP'])
df ['HP']=np.where(df['HP']<IrHP,IrHP,df['HP'])
df ['HP']=np.where(df['HP']>urHP,urHP,df['HP'])

IrCylinders,urCylinders=remove_outlier(df['Cylinders'])
df ['Cylinders']=np.where(df['Cylinders']<IrCylinders,IrCylinders,df['Cylinders'])
df ['Cylinders']=np.where(df['Cylinders']>urCylinders,urCylinders,df['Cylinders'])

df. shape
```

```python
median1 = df['HP'].median()
median2 = df['Cylinders'].median()

df['HP'] = df['HP'].fillna(median1)
df['Cylinders'] = df['Cylinders'].fillna(median2)
```
